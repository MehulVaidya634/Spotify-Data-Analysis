```python
# Top 10 most dancable songs.
df[['name', 'artists', 'album', 'danceability', 'year']].sort_values(by = 'danceability', ascending = False).head(10)
```

```python
# What is the average, median and mode duration in minutes of songs.
print('The Average duration in mins: ',df['duration_ms'].mean()/60000)
print('The Average duration in mins: ',df['duration_ms'].median()/60000)
print('The Average duration in mins: ',df['duration_ms'].mode()/60000)
```
